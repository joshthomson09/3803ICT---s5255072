```python
import requests, csv
from bs4 import BeautifulSoup

def get_movies(url):
    response = requests.get(url)
    soup_imdb = BeautifulSoup(response.text, 'html.parser')

    all_movies = soup_imdb.find_all("td", {"class":"titleColumn"})

    # get movie title :
    titles = [movie.find("a").get_text() for movie in all_movies]

    # get ranking :
    rankings = [movie.find("span").get_text()[:-1] for movie in all_movies]

    # get release year :
    list_year = [movie.find("span", {"class":"secondaryInfo"}).get_text() for movie in all_movies]
    years = [element.strip("()") for element in list_year]

    # get casting :
    castings = []
    for movie in all_movies:
        actors = [actor.get_text() for actor in movie.find_all("a")[1:]]
        castings.append(", ".join(actors))

    # get rating :
    list_ratings = soup_imdb.select(".ratingColumn")
    ratings = [rating.get_text(strip=True) for rating in list_ratings]

    mega_list = list(zip(titles, rankings, years, castings, ratings))
    mega_list = [list(elt) for elt in mega_list]

    # Write csv
    str_lst = [f"{mega_list[i][1]} / {mega_list[i][0]} ({mega_list[i][2]}) / Starring: {mega_list[i][3]} / Rating: {mega_list[i][4]}" for i, val in enumerate(mega_list)]

    res = [elt.split("/") for elt in str_lst]

    with open("imdb_top_250.csv", "w", newline='') as f:
        writer = csv.writer(f, delimiter=",")
        writer.writerow(['Ranking', 'Title', 'Release Year', 'Cast', 'Rating'])
        for row in res:
            writer.writerow([col.strip() for col in row])

get_movies("https://www.imdb.com/chart/top")
```

```python
def main():
    url = "https://www.imdb.com/chart/top"
    get_movies(url)
    

if __name__ == "__main__":
    main()
```
